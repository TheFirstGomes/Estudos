```python
import pandas as pd
import random
```

```python
culturas = ['Soja', 'Milho', 'Café', 'Arroz', 'Cana-de-Açucar', 'Trigo', 'Feijão']
produtividade = ['Baixa', 'Média', 'Alta', 'Muito Alta']

data = {
    'num_trabalhadores': [random.randint(10, 45) for _ in range(100)], 
    'producao_toneladas': [round(random.uniform(400, 2500), 2) for _ in range(100)], 
    'cultura':[random.choice(culturas) for _ in range(100)], 
    'classificacao_produtividade': [random.choice(produtividade) for _ in range(100)]
}
```

```python
df = pd.DataFrame(data)
df
```
